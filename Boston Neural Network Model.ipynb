```python
# Boston Housing Data set
# Regression Model with help of Keras Library
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
from sklearn.datasets import load_boston
```

```python
Boston = load_boston()
```

```python
Boston
```

```python
# Convert the data into pandas dataframe
df = pandas.DataFrame(data= Boston.data, columns= Boston.feature_names)
```

```python
df['Target'] = Boston.target
```

```python
df.head()
```

```python
df.shape
```

```python
# split into input and output variables
X= df.iloc[:,0:13]
Y= df.iloc[:, 13]
```

```python
def model():
    # create model
    model=Sequential()
    model.add(Dense(20,input_dim=13, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# evaluate model with standardized dataset
estimators=[]
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=model , epochs = 150 , batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state= seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print('Model : %.2f (%.2f) MSE' %(results.mean(), results.std()))
                    
```
